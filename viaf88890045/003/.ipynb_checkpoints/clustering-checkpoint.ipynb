```python
import re

firstparatexts = {}
entryheads = {}
wordlist = {}
wordentries = {}
entrywords = {}
entryheadwords = {}
headplusfirstpara = {}
entries = 0
tfidfvals = {}
firstpara = 0
def getentries(fname):
    global entries
    f = open(fname)
    curentryid = ''
    for l in f:
        m = re.search('entry" xml:id="([^"]+)',l)
        if(m):
            curentryid = m[1]
            entries = entries + 1
            entrywords[curentryid] = {}
            firstpara = 1
            continue
        if(not curentryid):
            continue

        if(re.search('<p>',l)):
            if(firstpara):
                workl = l.lower()
                workl = re.sub('<bibl n="([^"]+)">[^<]+</bibl>','\g<1>',workl)
                workl = re.sub('<[^>]+>',' ',workl)
                firstparatexts[curentryid] = workl
                workl = re.sub('[,;]','',workl)
                workl = re.sub("'s$",'',workl)
                workl = re.sub("s'$",'s',workl)
                workl = re.sub("([a-zA-Z])'([a-zA-Z])",'\g<1>\g<2>',workl)
                for foo in workl.lower().split():
                    if(foo in headplusfirstpara):
                        headplusfirstpara[foo].append(curentryid) 
                    else:
                        headplusfirstpara[foo] =[curentryid]
                    
            firstpara = 0
        
        m = re.search('<head>(.+)</head>',l)
        if(m and not curentryid in entryheads):
            entryheads[curentryid] = m[1]
            curtext = re.sub('\.</genName>','',m[1])
            curtext = re.sub('<[^>]+>','',curtext)
            curtext = re.sub('[\,;]+',' ',curtext)
            curtext = re.sub("'s$",'',curtext)
            curtext = re.sub("s'$",'s',curtext)
            curtext = re.sub("([a-zA-Z])'([a-zA-Z])",'\g<1>\g<2>',curtext)
            words = curtext.lower().split()
            for foo in words:
                if(foo in headplusfirstpara):
                    headplusfirstpara[foo].append(curentryid) 
                else:
                    headplusfirstpara[foo] =[curentryid]
                if(foo in entryheadwords):
                    entryheadwords[foo].append(curentryid) 
                else:
                    entryheadwords[foo] =[curentryid]
            
            
        workl = re.sub('<bibl[^>]+>[^<]+</bibl>',' ',l)
        workl = re.sub('<title[^>]+>[^<]+</title>',' ',workl)
        workl = re.sub('<[^>]+>',' ',workl)
        workl = re.sub('[\.;,!"()]', ' ',workl)
        workl = re.sub('\'s\\b','',workl)
        workl = re.sub('[0-9]+',' ',workl)
        words = workl.split()
        for word in words:
            if(word in entrywords[curentryid]):
                entrywords[curentryid][word] = entrywords[curentryid][word] + 1
            else:
                entrywords[curentryid][word] = 1
            if(word in wordlist):
                wordlist[word] = wordlist[word] + 1
                worklist = wordentries[word]
                if(curentryid in worklist):
                    worklist[curentryid] = worklist[curentryid] + 1
                else:
                    worklist[curentryid] =  1
                wordentries[word] = worklist
                    
            else:
                wordlist[word] =  1
                worklist = {}
                worklist[curentryid] = 1
                wordentries[word] = worklist
    
    f.close()
    
getentries('smithbio-work.xml')
print('1',entries)
getentries('../002/smithgeo-work.xml')
print('2',entries)

```

```python
import re
entrydates = {}

def prepfile(fname):
    f = open(fname)
    curentryid = ''
    firstpara = 0
    for l in f:
        if(re.search('id="porcifera-geo',l)):
            print('hit',l)
        m = re.search('entry" xml:id="([^"]+)"',l)
        if(m):
            curentryid = m[1]
            firstpara = 1
            entrydates[curentryid] = {}
        if(not curentryid):
            continue
        if(re.search('<p>',l)):
            if(firstpara):
                workl = l.lower()
                workl = re.sub('<bibl n="([^"]+)">[^<]+</bibl>','\g<1>',workl)
                workl = re.sub('<[^>]+>',' ',workl)
                firstparatexts[curentryid] = workl            
            firstpara = 0

        matchs = '<date when-custom="([^"]+)"'
        while(re.search(matchs,l)):
            m = re.search(matchs,l)
            curdate = m[1]
            nmatchs = '([\-]*[0-9]+)-([\-]*[0-9]+)'
            if(re.search(nmatchs,curdate)):
                p = re.search(nmatchs,curdate)
                #print('compdate',p[1],p[2])
                startdate = int(p[1])
                enddate = int(p[2])
            else:
                startdate = int(curdate)
                enddate = startdate + 1
            for i in range(startdate,enddate):
                if(curdate in entrydates[curentryid]):
                    entrydates[curentryid][i] = entrydates[curentryid][i] + 1
                else:
                    entrydates[curentryid][i] =  1
            l  = re.sub(matchs,' ',l,1)
prepfile('smithbio-work.xml')         
prepfile('../002/smithgeo-work.xml')         
print('done1')
```

```python
import re

friendlist = {}
frienddates = {}

contemporarylist = {}
contemporarydates = {}

siblinglist = {}
siblingdates = {}

spouselist = {}
spousedates = {}

consullist = {}
consuldates = {}

teacherlist = {}
teacherdates = {}

childlist = {}
childdates = {}

parentlist = {}
parentdates = {}

reignlist = {}
reigndates = {}

battlelist = {}
battledates = {}
warlist = {}
wardates = {}
warwithlist = {}
warwithdates = {}
siegelist = {}
curentryid = ''
lastdate = 0
siegedates = {}

def newentrel(ents,lastdate,entlist,entdates):

    if(ents in entlist):
        entlist[ents] = entlist[ents] + 1
    else:
        entlist[ents] = 1
        entdates[ents] = {}

    if(lastdate):
        workdict = entdates[ents]
        if(lastdate in workdict):
            workdict[lastdate] = workdict[lastdate] + 1
        else:
            workdict[lastdate] = 1
        entdates[ents] = workdict
                
    return(entlist,entdates)

namestring = '([A-Z][npt]*\.\s+[A-Z][a-z]+\s+[A-Z][a-z]+|[A-Z][npt]*\.\s+[A-Z][a-z]+|the\s+younger\s+[A-Z][a-z]+|[A-Z][a-z]+\s+the\s+younger|[A-Z][a-z]+\s+the\s+elder|the\s+elder\s+[A-Z][a-z]+|[A-Z][a-z]+\s+[A-Z][A-Za-z]+|[A-Z][a-z]+\s+the\s+[A-Za-z]+|[A-Z][a-z]+\s+of\s+[A-Za-z]+|[A-Z][a-z]+)'
def getchronphrases(fname):
    global childlist, childdates, parentlist, parentdates
    global reignlist, reigndates, warlist, wardates, warwithlist, warwithdates
    global battlelist, battledates, siegelist, siegedates
    global consullist, consuldates, teacherlist, teacherdates
    global siblinglist, siblingdates, spouselist, spousedates
    global contemporarylist, contemporarydates, friendlist, frienddates

    f = open(fname)
    
    friendstring = '\\b(friend\s+of\s+)' + namestring
    contemporarystring = '\\b(contemporary\s+of\s+)' + namestring
    teacherstring = '\\b(student\s+of\s+|disciple\s+of\s+)' + namestring
    parentstring = '\\b(son\s+of\s+|daughter\s+of\s+)' + namestring
    spousestring = '\\b(husband\s+of\s+|wife\s+of\s+)' + namestring
    siblingstring = '\\b(sister\s+of\s+|brother\s+of\s+)' + namestring
    childstring = '\\b(father\s+of\s+|mother\s+of\s+)' + namestring
    reignstring = '(the\s+reign\s+of\s+|the\s+emperor\s+)' + namestring
    warwithstring = '(war\s+with)\s+([A-Z][a-z]+|the\s+[A-Z][a-z]+)\\b'
    warstring = '(the)\s+([A-Z][a-z]+)\s+([Ww]ar)\\b'
    battlestring = '(defeated at|victory at|battle of|battle near|defeat at)\s+((mount)\s+[A-Z][a-z]+|[A-Z][a-z]+[\-\s]+[A-Z][a-z]+|[A-Z][a-zâ]+|the\s+([lL]ake|[Rr]iver)\s+[A-Z][a-z]+|the\s+[A-Z][a-z]+\s+([Gg]ate|[Bb]ridge|[Ff]orest|[Ll]ake|[Rr]iver|[Gg]ames|rock|[A-Z][a-z]+)|the\s+[A-Z][a-z]+)'
    siegestring = '(siege\s+of|siege\s+to)\s+((mount)\s+[A-Z][a-z]+|[A-Z][a-z]+[\-\s]+[A-Z][a-z]+|[A-Z][a-zâ]+|the\s+([lL]ake|[Rr]iver)\s+[A-Z][a-z]+|the\s+[A-Z][a-z]+\s+([Gg]ate|[Bb]ridge|[Ff]orest|[Ll]ake|[Rr]iver|[Gg]ames|rock|[A-Z][a-z]+)|the\s+[A-Z][a-z]+)'
    consulstring = '(\\bconsulship\s+of\s+)' + namestring
    for l in f:
        m = re.search('entry" xml:id="([^"]+)"',l)
        if(m):
            curentryid = m[1]
            lastdate = 0
          
        if(re.search('<p>',l)):
            lastdate = 0
            
        matchs = '<date when-custom="([^"]+)"'
        while(re.search(matchs,l)):
            m = re.search(matchs,l)
            
            lastdate = int(re.sub('([0-9])-.+','\g<1>',m[1]))
            l = re.sub(matchs,' ',l,1)
            
        m = re.search(friendstring,l)
        if(m):
            friendlist,frienddates = newentrel(m[2],lastdate,friendlist,frienddates)
            
        m = re.search(contemporarystring,l)
        if(m):
            contemporarylist,contemporarydates = newentrel(m[2],lastdate,contemporarylist,contemporarydates)
            
        m = re.search(spousestring,l)
        if(m):
            spouselist,spousedates = newentrel(m[2],lastdate,spouselist,spousedates)
            
        m = re.search(teacherstring,l)
        if(m):
            teacherlist,teacherdates = newentrel(m[2],lastdate,teacherlist,teacherdates)
            
        m = re.search(siblingstring,l)
        if(m):
            siblinglist,siblingdates = newentrel(m[2],lastdate,siblinglist,siblingdates)
            
        m = re.search(childstring,l)
        if(m):
            childlist,childdates = newentrel(m[2],lastdate,childlist,childdates)
            
        m = re.search(consulstring,l)
        if(m):
            consullist,consuldates = newentrel(m[2],lastdate,consullist,consuldates)

                
            
        m = re.search(parentstring,l)
        if(m):
            parentlist,parentdates = newentrel(m[2],lastdate,parentlist,parentdates)

            
        m = re.search(reignstring,l)
        if(m):
            reignlist,reigndates = newentrel(m[2],lastdate,reignlist,reigndates)
                
        m = re.search(warstring,l)
        if(m):
            warlist,wardates = newentrel(m[2],lastdate,warlist,wardates)

                
        m = re.search(warwithstring,l)
        if(m):
            warwithlist,warwithdates = newentrel(m[2],lastdate,warwithlist,warwithdates)
                
        m = re.search(siegestring,l)
        if(m):
            siegelist,siegedates = newentrel(m[2],lastdate,siegelist,siegedates)
                
                
        m = re.search(battlestring,l)
        if(m):
            works = m[2]
            if(re.search(' [Gg]ames',works)):
                continue
            battlelist,battledates = newentrel(works,lastdate,battlelist,battledates)

    
    f.close()

getchronphrases('smithbio-work.xml')   
getchronphrases('../002/smithgeo-work.xml')   
print('friendlist',len(friendlist))
print('contemporarylist',len(contemporarylist))
print('spouselist',len(spouselist))
print('siblinglist',len(siblinglist))
print('teacherlist',len(teacherlist))
print('consullist',len(consullist))
print('parentlist',len(parentlist))
print('childlist',len(childlist))
print('reigns',len(reignlist))
print('sieges',len(siegelist))
print('wars',len(warlist))
print('warwiths',len(warwithlist))
print('battles',len(battlelist))

```

```python
import re
smithngrams = {}

def preprocsent(workl):
    
    workl = re.sub('<abbr[/]*>','',workl)
    workl = re.sub('<bibl[^>]+>[^<]+</bibl>',' ',workl)
    workl = re.sub('<title[^>]+>[^<]+</bibl>',' ',workl)
    workl = re.sub('<date[^>]+>[^<]+</date>',' ',workl)
    workl = re.sub('[\(\)]',' ; ',workl)
        
    workl = re.sub('<[^>]+>',' ',workl)
    workl = re.sub('^[0-9\.\s]+','',workl)
    words = workl.split()
    if(len(words) == 0):
        return('')
    if(words[0].lower() in wordlist and words[0] in wordlist):
        if( wordlist[words[0].lower()] > wordlist[words[0]]*2):
            letuc = workl[0]
            letlc = workl[0].lower()
            workl = re.sub('^.',letlc,workl)
    
    workl = re.sub('([A-Z][a-z]+)\s+(of|the)','\g<1>-\g<2>',workl)
    workl = re.sub('([A-Z][a-z]+)\s+(of\s+the)','\g<1>-of-the',workl)
    return(workl)

def getnames(fname):
    inheader = 1
    curentryid = ''
    f = open(fname)
    for l in f:
        if(re.search('</teiHeader>',l)):
            
            inheader = 0
            continue
        if(inheader):
            continue
            
            
        m = re.search('entry" xml:id="([^"]+)',l)
        if(m):
            curentryid = m[1]
        if(not curentryid):
            continue
        
        if(not re.search('(<p\\b|<s\\b)',l)):
            continue
        workl = preprocsent(l)
        words = workl.split()
        nwords = len(words)
        for i in range(0,nwords):

            if(words[i].islower()):
                continue
            k = i
            ngram = ''
            while(k<nwords):
                if(words[k][0].isupper()):
                    if(ngram):
                        ngram = ngram + ' '
                    ngram = ngram + words[k]
                    if(re.search('[,;]$',ngram)):
                        ngram = re.sub('[,;]$','',ngram)
                        k = nwords
                    ngram = re.sub('[\-\)]+$','',ngram)
                    ngram = re.sub('(;|\.|\'s|\')$','',ngram)
                    if(ngram in smithngrams):
                        smithngrams[ngram] = smithngrams[ngram] + 1
                    else:
                        smithngrams[ngram] = 1
                else:
                    break
                k = k + 1
            if(not ngram):
                continue
    
    f.close()
    
def getentrynames(fname,searchid):
    f = open(fname)
    j = 0
    inheader = 1
    curentryid = ''
    for l in f:
        l = re.sub('\.\s*($|\(|</p>|\[)',' \g<1>',l)
        if(re.search('</teiHeader>',l)):
            
            inheader = 0
            continue
        if(inheader):
            continue
            
        m = re.search('entry" xml:id="([^"]+)',l)
        if(m):
            curentryid = m[1]
        
        if(not re.search(searchid,curentryid)):
            continue
        if(not re.search('(<p\\b|<s\\b)',l)):
            continue
        j = j + 1
        if(j> 500):
            continue
        workl = preprocsent(l)
        
        
        words = workl.split()
        nwords = len(words)
        print()
        print(l)
        jumpn = -1
        for i in range(0,nwords):

            if(i < jumpn):
                continue
            k = i
            ngram = ''
            while(k<nwords):
                if(words[k][0].isupper()):
                    if(ngram):
                        ngram = ngram + ' '
                    ngram = ngram + words[k]
                    if(re.search('[,;]$',ngram)):
                        ngram = re.sub('[,;]$','',ngram)
                        k = nwords
                    k = k + 1
                    ngram = re.sub('[\-\)]+$','',ngram)
                    ngram = re.sub('(;|\'s|\')$','',ngram)
                    if(ngram):
                        if(not  ngram in smithngrams):
                            val = 0
                        else:
                            val = smithngrams[ngram]
                    if(len(ngram.split())==1 and ngram in wordentries):
                        val2 = len(wordentries[ngram])
                    else:
                        val2 = 'NA'
                    rval1,rval2 = getbestentry2(ngram)
                    if(len(rval1)< len(rval2)):
                        rlist = rval1
                    else:
                        rlist = rval2
                    print('maxngram',i,k,curentryid,val,val2,ngram,rlist)
                    if(len(rlist) == 1):
                        jumpn = i + len(ngram.split())
                        print('jump',jumpn)
                else:
                    break

            
    f.close()

getnames('smithbio-work.xml')
getnames('../002/smithgeo-work.xml')
i = 0

getentrynames('smithbio-work.xml','thucydides')
print('done')
```
